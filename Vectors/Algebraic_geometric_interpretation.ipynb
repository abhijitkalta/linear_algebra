```python
## Distributive property - a.(b + c) = a.b + a.c

#create random vectors
n = 10
rng = np.random.default_rng(42)
a = rng.random(n)
b = rng.random(n)
c = rng.random(n)

#results
res1 = np.dot(a, b + c)
res2 = np.dot(a, b) +  np.dot(a, c)
print([res1, res2])
```

```python
## Associative property - a.(b.c) != (a.b).c

#create random vectors
n = 5
rng = np.random.default_rng(42)
a = rng.random(n)
b = rng.random(n)
c = rng.random(n)

#results
dp1 = np.dot(b, c)
dp2 = np.dot(a, b)
res1 = np.dot(a, dp1)
res2 = np.dot(dp2, c)
print("Associative property doesnt hold:\n", [res1, res2])  #associative property doesn't hold for vectors
print("\n")

#special case when associative property hold
#a==b==c

res1 = np.dot(a, np.dot(a, a))
res2 = np.dot(np.dot(a, a), a)
print("Associative property holds when all the vectors are same:\n", [res1, res2])

#all are zero vectors
z = np.zeros(n)
res1 = np.dot(z, np.dot(z, z))
res2 = np.dot(np.dot(z, z), z)
print([res1, res2])
```

```python
##Dot product - geometric perspective

v1 = np.array([2, 4, -3,])
v2 = np.array([0, -3, -3,])

#compute the angle between two vectors
ang = np.arccos(np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2)))


#plot
fig = plt.figure()
axes = fig.add_subplot(projection = '3d')
axes.plot([0, v1[0]], [0, v1[1]], [0, v1[2]], 'b', label = 'v1')
axes.plot([0, v2[0]], [0, v2[1]], [0, v2[2]], 'r', label = 'v2')

fig.legend()
plt.axis((-6, 6, -6, 6))
fig.suptitle('Angle between them: {}'.format(ang))
plt.show()
```

```python
##Equivalnece of algebraic and geomtric dot product formulas

v1 = np.array([2, 4, -3,])
v2 = np.array([0, -3, -3,])

dp_a = np.dot(v1, v2)  #algebraic notation

dp_g = np.linalg.norm(v1) * np.linalg.norm(v2) * np.cos(ang)

print(dp_a, dp_g)
```

```python
##Cauchy-Schqarz inequality - |aTb| <= ||a||*||b||

a = np.random.randn(5)
b = np.random.randn(5)
c = np.random.randn(1) * a

#compute their dot products
aTb = np.dot(a, b)
aTc = np.dot(a, c)  #linearly dependent so cos is either -1 or 1

mag = np.linalg.norm(a) * np.linalg.norm(b)
magC = np.linalg.norm(a) * np.linalg.norm(c)

print(abs(aTb), mag)  #always less than equal
print(abs(aTc), magC)  #equal in case of cos -1 or 1
```

```python
##cross product

v1 = [-3, -2, 5]
v2 = [-4, -3, 0]

v3a = np.cross(v1, v2)

fig = plt.figure()
axes = fig.add_subplot(projection = '3d')

#draw plane
xx, yy = np.meshgrid(np.linspace(-10, 10, 10), np.linspace(-10, 10, 10))
z1 = (-v3a[0] * xx - v3a[1] * yy) / v3a[2]
axes.plot_surface(xx, yy, z1, cmap='winter')

axes.plot([0, v1[0]],[0, v1[1]], [0, v1[2]], 'k', lw = 5)
axes.plot([0, v2[0]],[0, v2[1]], [0, v2[2]], 'k', lw = 5)
axes.plot([0, v3a[0]],[0, v3a[1]], [0, v3a[2]], 'r')

axes.view_init(azim = 150, elev = 45)
plt.show()
```

```python
##Hermitian transpose

z = np.array([3 + 4j])

#magnitude
print(np.linalg.norm(z))
print(np.dot(z, z))
print(np.transpose(z)*z)

#hermitian transpose
print(np.transpose(z.conjugate()) * z)
print('')

#complex vector
v = np.array([3, 4j, 5 + 2j, 2 - 5j])
print(v)
print(v.T)
print(np.transpose(v))
print(np.transpose(np.conjugate(v)))
print(np.conjugate(v))
```

```python
##Span and subspace

s1 = np.array([1, 1, 0])
s2 = np.array([1, 7, 0])

v = np.array([1, 2, 0])
w = np.array([3, 2, 1])

fig = plt.figure()
axes = fig.add_subplot(projection = '3d')
axes.plot([0, s1[0]], [0, s1[1]], [0.1, s1[2] + 0.1], 'r', lw = 3)
axes.plot([0, s2[0]], [0, s2[1]], [0.1, s2[2] + 0.1], 'r', lw = 3)

axes.plot([0, v[0]], [0, v[1]], [0.1, v[2] + 0.1], 'g', lw = 3)
axes.plot([0, w[0]], [0, w[1]], [0, w[2]], 'b', lw = 3)

xx, yy = np.meshgrid(range(-15, 16), range(-15, 16))
cp = np.cross(s1, s2)
z1 = (-cp[0] * xx - cp[1] * yy)*1./cp[2]
axes.plot_surface(xx, yy, z1, color='yellow', alpha = 0.3)

plt.show()
```
